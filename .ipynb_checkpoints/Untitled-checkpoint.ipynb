```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('./jobs_clean.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize = (5,3))
plt.hist(df['Salary Estimate'], bins=10, color = '#403d58')
plt.title('Distribution of Salary Estimates')
plt.xlabel('Salary Estimate')
plt.ylabel('Frequency')
plt.ylim((0,1200))
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (5,3))
plt.hist(df['Company Age'], bins=10, color = '#403d58')
plt.title('Distribution of Company Ages')
plt.xlabel('Age')
plt.ylim((0,900))
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (5,3))
plt.hist(df['Rating'], bins=10, color = '#403d58')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.ylim((0,750))
plt.show()
```

```python
plt.figure(figsize = (5,3))
df.Seniority.value_counts().plot(kind = 'bar', color='#403d58')
plt.title('Seniority Level Counts')
plt.xlabel('Seniority')
plt.ylim((0,600))
plt.ylabel('Frequency')
plt.xticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (5, 3))
df.Title_Simp.value_counts().plot(kind = 'bar', color = '#403d58')
plt.xticks(rotation = 45)
plt.title('Counts of Job Roles')
plt.xlabel('Job Role')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
df.Remote_yn.value_counts().plot(kind = 'bar', color = '#403d58')
plt.xticks(rotation = 0)
plt.title('Counts of Remote Jobs')
plt.xlabel('Remote (Yes/No)')
plt.ylim((0,2000))
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (5,3))
df.Location.value_counts().head(10).plot(kind='bar', color = '#403d58')
plt.title('Counts of Job Locations')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.ylim((0,1000))
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (5,3))
df.Sector.value_counts().head(10).plot(kind='bar', color = '#403d58')
plt.title('Counts of Job Sectors')
plt.xlabel('Sector')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (5,3))
df.Industry.value_counts().head(10).plot(kind='bar', color = '#403d58')
plt.title('Counts of Job Industries')
plt.xlabel('Industry')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (5,3))
Rev_df = df[df.Revenue.str.contains('USD', na = False)]
Rev_df.Revenue.value_counts().head(10).plot(kind='bar', color = '#403d58')
plt.title('Counts of Company Revenues')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (5,3))
size_df = df[df.Size.str.contains('employees', case = False, na = False)]
size_df.Size.value_counts().head(10).plot(kind='bar', color = '#403d58')
plt.title('Counts of Company Sizes')
plt.xlabel('Size')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (5,3))
own_df = df[~df['Ownership Type'].str.contains('Unknown', na = False)]
own_df['Ownership Type'].value_counts().head(10).plot(kind='bar', color = '#403d58')
plt.title('Counts of Company Ownership Types')
plt.xlabel('Ownership')
plt.ylabel('Frequency')
plt.show()
```

```python
corr = df[['Salary Estimate', 'Rating', 'revenue_simp', 'size_simp', 'Company Age', 'Remote_yn',
       'python_yn', 'r_yn', 'sql_yn', 'pytorch_yn', 'tensorflow_yn', 'scikit-learn_yn',
       'spark_yn', 'deeplearning_yn', 'tableau_yn','powerbi_yn', 'excel_yn', 'aws_yn', 'azure_yn', 'googlecloud_yn']].corr()
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(corr, cmap = 'crest')
plt.title('Corplot between Rating, Revenue, Size and Age of Companies')
plt.show()
```

```python
skill_counts = df[['python_yn', 'r_yn', 'sas_yn', 'sql_yn', 'altair_yn', 'talend_yn',
       'alteryx_yn', 'pytorch_yn', 'tensorflow_yn', 'scikit-learn_yn',
       'spark_yn', 'mongodb_yn', 'mysql_yn', 'deeplearning_yn', 'tableau_yn', 'powerbi_yn', 
       'excel_yn', 'aws_yn', 'azure_yn', 'googlecloud_yn']].sum().sort_values(ascending = False)
skill_counts
plt.bar(skill_counts.index, skill_counts.values, color = '#403d58')
plt.xticks(rotation=90)
plt.title('Most In-Demand Data Science Skills')
plt.xlabel('Skill')
plt.ylabel('Counts')
plt.ylim(0, 1000)
plt.show()
```

```python
sal_job = pd.pivot_table(df, index = ['Title_Simp'], columns = ['Seniority'], values = 'Salary Estimate', aggfunc='mean').sort_values('senior', ascending = False)
sal_job.plot(kind = 'bar')
plt.xticks(rotation = 45)
plt.xlabel('Job Role')
plt.ylabel('Average Salary')
plt.title('Average Salaries per Job Role')
plt.show()
```

```python
df.query('Title_Simp == "Manager" and Seniority == "junior" and Title_Simp.notna() and Seniority.notna()').sort_values('Salary Estimate', ascending=False)
```

```python
top_sect = df.Sector.value_counts().head(10).index
top_sect_df = df[df.Sector.isin(top_sect)]
sal_sect = pd.pivot_table(top_sect_df, index = 'Sector', values = 'Salary Estimate', aggfunc = 'mean').sort_values('Salary Estimate', ascending = False)
sal_sect.plot(kind='bar')
plt.show()
```

```python

sect_job = pd.pivot_table(top_sect_df, index = 'Sector', columns = 'Title_Simp', aggfunc = 'size')
ax = sect_job.plot(kind = 'bar')
plt.xticks(rotation = 90)
plt.xlabel('Sector')
plt.ylabel('Count')
plt.title('Job Demands per Sector for Each Role')
ax.legend(title = 'Job Role')
plt.show()
```

```python
# have an option in the webapp to input the industry name and have this distribution pop up 
sect_job = pd.pivot_table(df[df['Sector'].str.contains('Media', na = False, case = False)], index = 'Sector', columns = 'Title_Simp', aggfunc = 'size')
ax = sect_job.plot(kind = 'bar')
plt.xticks(rotation = 0)
plt.xlabel('Sector')
plt.ylabel('Count')
plt.title('Job Demands in the Media Sector for Each Role')
ax.legend(title = 'Job Role', loc = 'upper right')
plt.show()
```

```python
top_ind = df.Industry.value_counts().head(10).index
top_ind_df = df[df.Industry.isin(top_ind)]
ind_job = pd.pivot_table(top_ind_df, index = 'Industry', columns = 'Title_Simp', aggfunc = 'size')
ax = ind_job.plot(kind = 'bar')
plt.xticks(rotation = 90)
plt.xlabel('Industry')
plt.ylabel('Count')
plt.title('Job Demands per Industry for Each Role')
ax.legend(title = 'Job Role', loc = 'upper left')
plt.show()
```

```python
sal_sect = pd.pivot_table(top_ind_df, index = 'Industry', values = 'Salary Estimate', aggfunc = 'mean').sort_values('Salary Estimate', ascending = False)
sal_sect.plot(kind='bar')
plt.show()
```

```python
# have an option in the webapp to input the industry name and have this distribution pop up 
ind_job = pd.pivot_table(df[df['Industry'].str.contains('Business consulting', na = False, case = False)], index = 'Industry', columns = 'Title_Simp', aggfunc = 'size')
ax = ind_job.plot(kind = 'bar')
plt.xticks(rotation = 0)
plt.xlabel('Industry')
plt.ylabel('Count')
plt.title('Job Demands in the Business Consulting Industry for Each Role')
ax.legend(title = 'Job Role', loc = 'upper right')
plt.show()
```

```python
sal_rem = pd.pivot_table(df, index = 'Remote_yn', values = 'Salary Estimate', aggfunc = 'mean').sort_values('Salary Estimate', ascending = False)
sal_rem.plot(kind='bar')
plt.show()
```
