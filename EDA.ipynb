```python
# Ensure 'Salary Estimate' is in a numeric format
df['Salary Estimate'] = pd.to_numeric(df['Salary Estimate'], errors='coerce')

# Group by 'Location' and calculate the average salary and the count of job postings
location_salary_stats = df.groupby('Location').agg(
    avg_salary=('Salary Estimate', 'mean'),
    job_count=('Location', 'size')
).reset_index()

# Optionally, filter out locations with fewer job postings (e.g., less than 10)
filtered_location_salary_stats = location_salary_stats[location_salary_stats['job_count'] >= 10]

# Sort the results by average salary in descending order
sorted_location_salary_stats = filtered_location_salary_stats.sort_values(by='avg_salary', ascending=False)

print(sorted_location_salary_stats.head(10))  # Display the top 10 locations
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# bar plot for average salary
ax1.bar(sorted_location_salary_stats['Location'], sorted_location_salary_stats['avg_salary'], color='b', alpha=0.7)
ax1.set_xlabel('Location')
ax1.set_ylabel('average salary', color='b')
ax1.tick_params(axis='x', rotation=90)

# line plot for job count
ax2 = ax1.twinx()
ax2.plot(sorted_location_salary_stats['Location'], sorted_location_salary_stats['job_count'], color='r', marker='o')
ax2.set_ylabel('job count', color='r')

plt.title('average salary and job count per region')
plt.show()
```

```python
# Group by 'Location' and calculate the average salary and the count of job postings
location_rating_stats = df.groupby('Location').agg(
    avg_rating=('Rating', 'mean'),
    job_count=('Location', 'size')
).reset_index()

# Optionally, filter out locations with fewer job postings (e.g., less than 10)
filtered_rating_location_stats = location_rating_stats[location_rating_stats['job_count'] >= 10]

# Sort the results by average salary in descending order
sorted_location_rating_stats = filtered_rating_location_stats.sort_values(by='avg_rating', ascending=False)

print(sorted_location_rating_stats.head(10))  # Display the top 10 locations
```

```python
# plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# bar plot for average salary
ax1.bar(sorted_location_rating_stats['Location'], sorted_location_rating_stats['avg_rating'], color='b', alpha=0.7)
ax1.set_xlabel('location')
ax1.set_ylabel('average rating', color='b')
ax1.tick_params(axis='x', rotation=90)

# line plot for job count
ax2 = ax1.twinx()
ax2.plot(sorted_location_rating_stats['Location'], sorted_location_rating_stats['job_count'], color='r', marker='o')
ax2.set_ylabel('job count', color='r')

plt.title('average rating and job count per region')
plt.show()
```

```python
skills_col = ['python_yn', 'r_yn', 'sas_yn', 'sql_yn', 'altair_yn', 'talend_yn',
       'alteryx_yn', 'pytorch_yn', 'tensorflow_yn', 'scikit-learn_yn',
       'spark_yn', 'mongodb_yn', 'mysql_yn', 'deeplearning_yn', 'tableau_yn',
       'powerbi_yn', 'excel_yn', 'aws_yn', 'azure_yn', 'googlecloud_yn']
Skill_Count = df[skills_col].sum().sort_values(ascending = False)

Skill_Count.plot(kind = 'bar')
plt.title('Most Popular Skills')
plt.xlabel('Skill')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'Salary Estimate' is your salary column
average_salaries = {}

# Loop through each skill column
for skill in skills_col:
    # Filter the rows where the skill is present (== 1)
    skill_df = df[df[skill] == 1]
    
    # Calculate the average salary for this skill
    average_salary = skill_df['Salary Estimate'].mean()
    
    # Store the result in the dictionary
    average_salaries[skill] = average_salary

# Convert the dictionary to a pandas Series and sort by salary
average_salaries = pd.Series(average_salaries).sort_values(ascending=False)

# Plot the bar plot
average_salaries.plot(kind='bar', figsize=(10, 6))
plt.title('Average Salary by Skill')
plt.xlabel('Skill')
plt.ylabel('Average Salary')
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

# Step 1: Combine all job descriptions into a single string
text = ' '.join(df['Job Description'].dropna().tolist())

# Step 2: Define data science related stopwords
non_data_science_words = set(STOPWORDS)
additional_stopwords = {
    'the', 'to', 'and', 'a', 'of', 'in', 'is', 'for', 'with', 'on', 'that', 'as', 'at', 'be', 
    'are', 'or', 'an', 'we', 'will', 'by', 'this', 'it', 'from', 'you', 'your', 'can', 'our', 
    'which', 'but', 'have', 'has', 'more', 'not', 'all', 'if', 'these', 'they', 'also', 'their',
    'may', 'should', 'must', 'one', 'new', 'who', 'how', 'when', 'us', 'were', 'was', 'been', 'do', 'please', 'working', 'pension', 'will', 'u', 'g'
}

# Add additional stopwords to the default stopwords
non_data_science_words.update(additional_stopwords)

# Step 3: Create the word cloud
wordcloud = WordCloud(
    width=1920,
    height=1080,
    background_color='black',
    stopwords=non_data_science_words,
    collocations=False
).generate(text)

# Step 4: Plot the word cloud
plt.figure(figsize=(19.2, 10.8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
